```python
from PIL import Image
import io
from google.colab import files
import IPython.display as display

# Upload image
uploaded = files.upload()

# Get image path
image_path = list(uploaded.keys())[0]

# Input encryption key (between 0â€“255)
key = 240

# Encrypt or decrypt function
def encrypt_decrypt_image(path, key, mode="encrypt"):
    image = Image.open(path).convert("RGB")
    pixels = image.load()
    width, height = image.size

    for y in range(height):
        for x in range(width):
            r, g, b = pixels[x, y]
            # XOR encryption
            new_pixel = (r ^ key, g ^ key, b ^ key)
            pixels[x, y] = new_pixel

    return image

# Encrypt the image
encrypted_img = encrypt_decrypt_image(image_path, key, "encrypt")
encrypted_img_path = "encrypted_image.png"
encrypted_img.save(encrypted_img_path)

# Decrypt the image
decrypted_img = encrypt_decrypt_image(encrypted_img_path, key, "decrypt")
decrypted_img_path = "decrypted_image.png"
decrypted_img.save(decrypted_img_path)

# Display images
print("Original Image:")
display.display(Image.open(image_path))

print("Encrypted Image:")
display.display(Image.open(encrypted_img_path))

print("Decrypted Image:")
display.display(Image.open(decrypted_img_path))
```
